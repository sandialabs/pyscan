# a02 Example - Multi Live Plot

```python
import pyscan as ps
import numpy as np
import matplotlib.pyplot as plt
```

## Setup devices

```python
devices = ps.ItemAttribute()

devices.v1 = ps.TestVoltage() # Device 1
devices.v2 = ps.TestVoltage() # Device 2
devices.v3 = ps.TestVoltage() # Device 3
```

## Define a measure function

```python
def get_voltage_data(expt):
    """
    Reads the voltage from v1, v2, and v3 devices. Also adds a calculated value vsum.
    """

    runinfo = expt.runinfo
    devices = expt.devices

    d = ps.ItemAttribute()

    d.v1_readout = devices.v1.voltage/4
    d.v2_readout = devices.v2.voltage/2
    d.v3_readout = devices.v3.voltage

    d.vsum = d.v1_readout + d.v2_readout + d.v3_readout
    
    
    return d
```

## Define Plot Functions

```python
# define plotting functions, each taking expt and ax as arguments
def plot1(expt, ax):
    ax.plot(expt.v1_voltage, expt.v1_readout)
    
    ax.set_xlabel('V1 voltage')
    ax.set_ylabel('V1 readout')
    
    ax.set_title('Plot 1')

def plot2(expt, ax):
    ax.plot(expt.v1_voltage, expt.v1_readout, zs=expt.vsum)

    ax.set_xlabel('V1 voltage')
    ax.set_ylabel('V1 readout')
    ax.set_zlabel('vsum')
    
    ax.set_title('Plot 2')

def plot3(expt, ax):
    ax.plot(expt.v1_voltage, expt.vsum)

    ax.set_xlabel('V1 voltage')
    ax.set_ylabel('vsum')
    
    ax.set_title('Plot 3')
```

## Create the multiplot function

```python
live_multiplot = ps.multi_live_plot([plot1, plot2, plot3], subplot_locators=[221,222,212], projections=[None,'3d',None])
```

## Experiments

```python
# Create RunInfo instance and set scan0 to PropertyScan
runinfo = ps.RunInfo()
runinfo.scan0 = ps.PropertyScan({'v1': ps.drange(0, 0.1, 1), 'v2': ps.drange(0, 1, 10)}, prop='voltage', dt=1)

# Set RunInfo measure_function (remember, it takes a Experiment object as a parameter and 
# returns an ItemAttribute containing data). This function is run every time the experiment iterates one scan variable.
runinfo.measure_function = get_voltage_data

# Create a Experiment class with the RunInfo and Devices just created
expt = ps.Experiment(runinfo, devices, time=True)
expt.start_thread()
```

```python
try:
    live_multiplot(expt)
except KeyboardInterrupt:
    expt.stop()
```
